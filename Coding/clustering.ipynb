```python
import pandas as pd
import hvplot.pandas
from pathlib import Path
import matplotlib as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
df = pd.read_csv(
    Path('Data/kc_house_data.csv'),
    index_col='date',
    infer_datetime_format=True,
    parse_dates=True
)

df.head()
```

```python
df = df.drop(columns=['id'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Scale the data using Standard Scaler from sklearn
scaled_data = StandardScaler().fit_transform(df)
```

```python
# Create new dataframe with scaled data
df_scaled = pd.DataFrame(scaled_data, columns=['price','bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition','grade','sqft_above','sqft_basement','yr_built','yr_renovated','zipcode','lat','long','sqft_living15','sqft_lot15'])
df_scaled.head()
```

```python
k = list(range(1,11))
```

```python
inertia = []
```

```python
for i in k:
    model = KMeans(n_clusters=i, random_state=0)
    model.fit(df)
    inertia.append(model.inertia_)
```

```python
elbow_data = {
    'k': k,
    'inertia': inertia
}
df_elbow = pd.DataFrame(elbow_data)
```

```python
df_elbow.hvplot.line(x='k', y='inertia', title='elbow curve', xticks=k)
```

```python
model = KMeans(n_clusters=3)
model.fit(df_scaled)
```

```python
clusters = model.predict(df_scaled)
print(clusters)
```

```python
# Create a copy of the dataframe
df_scaled_predictions = df_scaled.copy()
df_scaled_predictions['cluster'] = clusters
df_scaled_predictions.head()
```

```python
# Use PCA (principal component analysis) to reduce the dataset to two features.
pca = PCA(n_components=2)
pca_data = pca.fit_transform(df_scaled)

# Display the sample data
pca_data[:5]
```

```python
# Measure the amount of variance in the Principal Components
pca.explained_variance_ratio_
```

```python
# This shows that our two principle components together only contain about 42.5% of the original information
```

```python
# Create a DataFrame with the PCA data
df_pca = pd.DataFrame(pca_data, columns=['PC1','PC2'])
df_pca.head()
```

```python
# Initialize the KMeans model
model = KMeans(n_clusters=3)
```

```python
# Fit the model
model.fit(df_pca)
```

```python
# Predict the clusters
clusters = model.predict(df_pca)
```

```python
# Create copy of the original DataFrame and add new column with predicted clusters
df_pca_predictions = df_pca.copy()
df_pca_predictions['clusters'] = clusters
df_pca_predictions.head()
```

```python
# Vizualize the clusters
df_pca_predictions.hvplot.scatter(x='PC1',y='PC2',by='clusters')
```
